## **Linear Regression**

Generating some linear looking data

```python
import numpy as np
np.random.seed(42) # to make this code example reproductible
m = 100 # m is the number of instances
X = 2*np.random.rand(m, 1) # column vector
y = 4 + 3 * X + np.random.randn(m,1) # column vector
```

Lets displaying this dataset using matplotlib

```python
from matplotlib import pyplot as plt
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```

Now firstly generating `X_b` from `X`, because at all `X0` we put 1<br>
Then evaluating best theta from upper data.

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X)
best_theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
best_theta
```

Now we can make predictions

```python
X_new = np.array([[0],[2]])
X_new_b = add_dummy_feature(X_new)
y_predict = X_new_b @ best_theta
y_predict
```

Now plotting this model's pediction

```python
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.scatter(X, y, label="Datapoints", s=5)
plt.legend()
plt.grid()
plt.show()
```

Performing Linear Regression with **Scikit-Learn**

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(X_new)
```
